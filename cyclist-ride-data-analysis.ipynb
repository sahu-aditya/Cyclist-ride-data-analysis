```r
#loading require packages
library(tidyverse)
library(lubridate)
library(ggplot2)
```

```r
#creating data frame to store the data by each month
march_21      <-read_csv("../input/cyclist-ride-data/202103-divvy-tripdata.csv")
april_21      <-read_csv("../input/cyclist-ride-data/202104-divvy-tripdata.csv")
may_21        <-read_csv("../input/cyclist-ride-data/202105-divvy-tripdata.csv")
june_21       <-read_csv("../input/cyclist-ride-data/202106-divvy-tripdata.csv")
july_21       <-read_csv("../input/cyclist-ride-data/202107-divvy-tripdata.csv")
august_21     <-read_csv("../input/cyclist-ride-data/202108-divvy-tripdata.csv")
september_21  <-read_csv("../input/cyclist-ride-data/202109-divvy-tripdata.csv")
october_21    <-read_csv("../input/cyclist-ride-data/202110-divvy-tripdata.csv")
november_21   <-read_csv("../input/cyclist-ride-data/202111-divvy-tripdata.csv")
december_21   <-read_csv("../input/cyclist-ride-data/202112-divvy-tripdata.csv")
january_22    <-read_csv("../input/cyclist-ride-data/202201-divvy-tripdata.csv")
february_22   <-read_csv("../input/cyclist-ride-data/202202-divvy-tripdata.csv")
```

```r
# checking the column name of each data frame to check the consistency
colnames(march_21)
colnames(april_21)
colnames(may_21)
colnames(june_21)
colnames(july_21)
colnames(august_21)
colnames(september_21)
colnames(october_21)
colnames(november_21)
colnames(december_21)
colnames(january_22)
colnames(february_22)
```

```r
#structure of data frame incluse data type, number of column, some value of each column
str(march_21)
str(april_21)
str(may_21)
str(june_21)
str(july_21)
str(august_21)
str(september_21)
str(october_21)
str(november_21)
str(december_21)
str(january_22)
str(february_22)
```

```r
# combining 12 months data frame into one data frame
all_trips <- bind_rows(march_21,april_21,may_21,june_21,july_21,august_21,september_21,october_21,november_21,december_21,january_22,february_22)
```

```r
colnames(all_trips)  #List of column names
nrow(all_trips)  #How many rows are in data frame?
dim(all_trips)  #Dimensions of the data frame?
head(all_trips)  #See the first 6 rows of data frame.  Also tail(all_trips)
str(all_trips)  #See list of columns and data types (numeric, character, etc)
summary(all_trips)  #Statistical summary of data. Mainly for numerics
```

```r
#some records of member_casual column have subscribe for member and customer for casual, making them to member and casual for consistency
all_trips <-  all_trips %>% 
  mutate(member_casual = recode(member_casual
                           ,"Subscriber" = "member"
                           ,"Customer" = "casual"))
```

```r
#adding date column extracted from started_at
all_trips$date <- as.Date(all_trips$started_at)
```

```r
#adding month column extracted from started_at
all_trips$month <- format(as.Date(all_trips$date),"%m")
```

```r
#similary adding day, year and days_of_weel column extracted from started_at
all_trips$day <- format(as.Date(all_trips$date),"%d")
all_trips$year <- format(as.Date(all_trips$date),"%y")
all_trips$days_of_week <- format(as.Date(all_trips$date),"%A")
```

```r
#colnames(all_trips)
head(all_trips)
```

```r
#calculating the lenth of ride
all_trips$ride_length <-difftime(all_trips$ended_at,all_trips$started_at)
```

```r
#converting the data type of ride_length to numeric for easy in further calculation
all_trips$ride_length <- as.numeric(as.character(all_trips$ride_length))
```

```r
#check the data type
is.numeric(all_trips$ride_length)
```

```r
#sometime the bike hasbeen in the service centre for maintainence and since those data here is not require. So removing them
all_trips_v2 <- all_trips[!(all_trips$start_station_name=="HQ QR" | all_trips$ride_length <0),]
```

```r
#Data contain sum na value
sum(is.na(all_trips_v2$ride_length))
```

```r
#removing na values
all_trips_v2 <- all_trips_v2 %>% na.omit(all_trips_v2$ride_length)
```

**Analyse**

Actual analysis on cleand data done here like finding the summerise aggregate results which may include mean, mode , median, max, min, average, sd, covariance, etc. and arrive to answer which analyst were seeking. and in last store in a new file to make the visualization for sharing phas.

```r
#calculation of mean, median,max and min
mean(all_trips_v2$ride_length)
median(all_trips_v2$ride_length)
max(all_trips_v2$ride_length)
min(all_trips_v2$ride_length)
```

```r
#summary of uncleand and clean data
summary(all_trips$ride_length)
summary(all_trips_v2$ride_length)
```

```r
#Finding the summary for individiual for member and casual
aggregate(all_trips_v2$ride_length~all_trips_v2$member_casual,FUN=mean )
aggregate(all_trips_v2$ride_length~all_trips_v2$member_casual,FUN=median)
aggregate(all_trips_v2$ride_length~all_trips_v2$member_casual,FUN=max)
aggregate(all_trips_v2$ride_length~all_trips_v2$member_casual,FUN=min)
```

```r
#Finding the summary for individiual for member and casual oncluding the days of week
aggregate(all_trips_v2$ride_length~all_trips_v2$member_casual+all_trips_v2$days_of_week, FUN=mean)
```

```r
#above result in the order fashion
all_trips_v2$days_of_week <- ordered(all_trips_v2$days_of_week, levels=c("Sunday","Monday","Tuesday","Wednesday","Thursday","Friday","Saturday"))
```

```r
aggregate(all_trips_v2$ride_length~all_trips_v2$member_casual+all_trips_v2$days_of_week, FUN=mean)
```

**Share**

In the share phase the analyst make the reports, dashboard or presentation for his stakehoderth share his finding on the data in efficeint manner. beloweare some example for visualization for share to stakeholder.

After presentation the analyst take necessary feedback and make relevat changes. 

```r
#creating a visual bar graph of number of ride for eachday of week by casual and member.
all_trips_v2 %>% mutate(weekday=wday(started_at,label=TRUE))%>% 
group_by(member_casual,weekday)%>% summarise(number_of_rides=n(),average_duration=mean(ride_length)) %>%
arrange(member_casual,weekday)%>% 
ggplot(aes(x=weekday,y=number_of_rides,fill=member_casual))+geom_col(position="dodge")
```

```r
#bar graph for average durationby casual and member for each day of week.
all_trips_v2 %>% 
  mutate(weekday = wday(started_at, label = TRUE)) %>% 
  group_by(member_casual, weekday) %>% 
  summarise(number_of_rides = n()
            ,average_duration = mean(ride_length)) %>% 
  arrange(member_casual, weekday)  %>% 
  ggplot(aes(x = weekday, y = average_duration, fill = member_casual)) +
  geom_col(position = "dodge")
```
